## Improve data_frame columns

```python
dict_per_data = {0: "Name", 1: "Age", 2: "Gender"}
dict_date = {0: "Month_Day", 1: "Year"}

def divide_column(column, n_dict):
  data_base = []
  for _ in column:
    _ = _ + ","
    element = {}
    ch_list = []
    counter = 0
    for character in _:
      if character != ",":
        ch_list.append(character)
      else:
        data = "".join(ch_list).strip()
        if n_dict == 0:
          if counter < 3:
            element[dict_per_data[counter]] = data
            ch_list.clear()
            counter += 1
        elif n_dict == 1:
          if counter < 2:
            element[dict_date[counter]] = data
            ch_list.clear()
            counter += 1
    data_base.append(element)
  return data_base
```

```python
def extract_data(data_base, field):
  list = []
  for i in data_base:
    age = i.get(field, np.nan)
    list.append(age)
  return pd.Series(list)
```

```python
#Divide the first column in 3: "Name", "Age", "Sex"
personal_data = divide_column(df["Name, Age, Gender"], 0)
name = pd.Series([_['Name'] for _ in personal_data])
age = extract_data(personal_data,'Age')
gender = extract_data(personal_data, 'Gender')
```

```python
df_personal_data = pd.DataFrame({
    "Name": name,
    "Age": age,
    "Gender": gender 
})
df_personal_data
```

```python
#Divide date column.
date = divide_column(df['Date'], 1)
month_day = extract_data(date, 'Month_Day')
year = extract_data(date, 'Year')
```

```python
df_date = pd.DataFrame({
    'Month_Day': month_day,
    'Year': year,
})
df_date
```

```python
df_final = pd.concat([df_personal_data, df_date, df.loc[:, ["Species", "Location", "Latitude", "Longitude"]]], 
                     axis=1)
df_final
```

```python
df_final = df_final.dropna()
df_final.shape
```
